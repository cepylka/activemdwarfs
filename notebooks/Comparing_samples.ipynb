Now we can compare samples in terms of rotation periods and mass of stars. The stars in this sample are indicated by red dots and the [Medina et al. 2020](#Medina) sample stars are indicated by green dots. Two stars with the smallest masses, Kepler-446 and TRAPPIST-1, are indicated on the plot.

In the cell below, from the calculated quiescent luminosity of the sample stars in the Kepler bandpass, we obtain the ratio of energy produced by flares at maximum star-specific observed energy and corresponding rate to the quiescent star energy per year and the ratio of energy produced by flares at the energy cut-off log10E = 31.5 erg to the quiescent star energy per year

```python
all_flares = pandas.read_pickle("../data/all_flares.pkl")
print(len(all_flares))
all_flares = all_flares.query("`ed_corr`.notna()")
print(len(all_flares), all_flares["flare_erg_rec"].mean())
for name in ros.index:
    print(name)
    print(len(all_flares[all_flares['name']==name]))
noflareslist = ["Kepler-446",
"Kepler-233",
"TOI-1749",
"K2-240"]
for name in ros.index:
    if name not in noflareslist:
        ros.at[name,"luminosity quiscent"] = all_flares[all_flares['name']==name]['luminosity quiscent'].values[0]

        ros.at[name,"duration_at_max"] = all_flares[all_flares['flare_erg_rec']==all_flares["flare_erg_rec"].max()]['duration_corr'].values[0]
        ros.at[name,"energy_at_max"] = all_flares[all_flares['flare_erg_rec']==all_flares["flare_erg_rec"].max()]['flare_erg'].values[0]
duration_year = 365*24*60*60

ros["Qlum_star_year"] = ros["luminosity quiscent"] * duration_year
ros["R31lum_flar_year"] = ros["freq_from_powerlaw"] *4.38e31 * 365
ros["rate_flar_year"] =  ros["R31lum_flar_year"]/ros["Qlum_star_year"]
ros
```
